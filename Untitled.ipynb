```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://datos.gob.ar/dataset/produccion-exportaciones-bienes-valores-cantidad-empresas-por-sector-actividad'

# Hacemos una petición HTTP a la página
response = requests.get(url)

# Analizamos el contenido HTML de la página con BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Buscamos todos los elementos <a> que contienen un atributo href
links = soup.find_all('a', href=True)

# Guardamos e imprimimos los enlaces que terminan en .csv
csv_links = []
for link in links:
    if link['href'].endswith('.csv'):
        print(link['href'])
        csv_links.append(link['href'])
```

```python
# Filtramos los enlaces que contienen la palabra "total"
filtered_links = [link for link in csv_links if "total" not in link]

# Imprimimos los enlaces filtrados
for link in filtered_links:
    print(link)
```

```python
# Leemos solo el data set clae_6 (El más específico segun clae)
# Leemos el diccionario de claes

empresas = pd.read_csv('https://cdn.produccion.gob.ar/cdn-cep/datos-por-actividad/expo/empresas-expo/empresas_expo_privada_por_clae6.csv')
diccionario = pd.read_csv('https://cdn.produccion.gob.ar/cdn-cep/clae_agg.csv')
```

```python
#Unimos los datos

datos = empresas.merge(diccionario, on='clae6')

#Vemos el data frame resultante

datos.info()
```

```python
# Selecciona las columnas innecesarias
cols_to_drop = ['clae3','clae2','letra']

# Elimina las columnas seleccionadas del DataFrame
datos = datos.drop(cols_to_drop, axis=1)
```

```python
datos['fecha']
```

```python
#Transformarmos la columna fecha en datetime

datos['fecha'] = pd.to_datetime(datos['fecha'])

#La establecemos como indice
datos = datos.set_index('fecha')
```

```python
#Veamos estadísticas de la cantidad de empresas
datos['empresas'].describe()
```

```python
#Filtraremos los datos que tienen menos de 0 empresas

datos = datos[~datos['empresas'] < 0]
```

```python
datos['empresas'].describe()
```

```python
grouped = datos.groupby([datos.index.year, 'letra_desc'])['empresas'].mean().to_frame().reset_index()
```
