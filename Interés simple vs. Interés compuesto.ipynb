```python
# Importaci√≥n de librer√≠as

import numpy as np
import matplotlib.pyplot as plt

#Inputs del usuario
capital_inicial = int(input("Ingrese el capital inicial: $"))
tasa_interes_anual = float(input("Ingrese la tasa de inter√©s anual: "))
plazo = int(input("Ingrese el plazo en meses en el que va a establecer el plazo fijo: "))

#Tasa de inter√©s mensual
tasa_interes_mensual = np.round(tasa_interes_anual / 12, 2)

#C√°lculo de inter√©s simple
is_capital = []
is_intereses = []

for i in range(plazo):
    is_incremento = np.round((capital_inicial * tasa_interes_mensual) / 100, 2)
    
    is_intereses.append(is_incremento)
    is_capital.append(capital_inicial)
    
    is_nuevo_capital_inicial = np.round(capital_inicial + is_incremento, 2)
    


#C√°lculo inter√©s compuesto
ic_capital = []
ic_intereses = []

for i in range(plazo):
    incremento = np.round((capital_inicial * tasa_interes_mensual) / 100, 2)
    nuevo_capital_inicial = np.round(capital_inicial + incremento, 2)
    
    ic_intereses.append(incremento)
    ic_capital.append(nuevo_capital_inicial)
    
    capital_inicial = nuevo_capital_inicial

#Graficos
fig, ax1 = plt.subplots(figsize=(15,4))

ax1.plot(ic_intereses, color='g', label='Intereses ganados renovando el capital + los intereses ganados cada 30 d√≠as')

ax1.plot(is_intereses, color='red', label='Intereses ganados dejando el capital inicial durante 12 meses')

etiquetas_x = ['Mes {}'.format(i+1) for i in range(plazo)]
ax1.set_xticks(range(plazo))
ax1.set_xticklabels(etiquetas_x, rotation=45)

#Calculo de rangos
minimo = min(is_intereses)
maximo = max(ic_intereses)
rango = maximo - minimo
margen = 0.1 * rango
ax1.set_ylim(minimo - margen, maximo + margen)
#Graficamos
plt.legend(loc=2, fontsize=10)
plt.show()

#Resultados finales
is_resultado = sum(is_intereses)+is_capital[0]
ic_resultado = sum(ic_intereses)+is_capital[0]
porcentaje_adicional = (sum(ic_intereses) - sum(is_intereses)) / sum(is_intereses) * 100

# Impresi√≥n de los resultados
print(f'El resultado bruto dejando el capital inicial en plazo fijo durante {len(is_intereses)} meses (inter√©s simple),es de ${np.round(is_resultado,2)} con una ganancia neta de $ {np.round(sum(is_intereses),2)}\n')
print(f'El resultado bruto renovando el capital inicial cada 30 d√≠as {len(ic_intereses)} meses (inter√©s compuesto) es de ${np.round(ic_resultado,2)} con una ganancia neta de $ {np.round(sum(ic_intereses),2)}.\n')
print(f'Esto equivale a una ganancia adicional del {porcentaje_adicional:.2f}% al renovar el capital cada 30 d√≠as en lugar de dejarlo en plazo fijo.')
```
